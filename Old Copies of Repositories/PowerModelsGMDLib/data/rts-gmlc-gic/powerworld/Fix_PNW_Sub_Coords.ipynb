```python
from esa import SAW
import os
```

```python
pwb_path = os.path.join(os.getcwd(), 'RTS-GMLC-GIC-PNW.pwb')
saw = SAW(pwb_path)
```

```python
(busses := saw.GetParametersMultipleElement(ObjectType='bus', ParamList=['BusNum','SubNum','Longitude','Latitude'])).head(20)
```

```python
busses[busses.SubNum == 1]
```

```python
(sub_coords := busses.groupby('SubNum')[['Longitude', 'Latitude']].mean()).head()
```

```python
sub_coords.reset_index(inplace=True)
sub_coords.head()
```

```python
# Old sub locations centered around OR
(subs := saw.GetParametersMultipleElement(ObjectType='substation', ParamList=['SubNum','Longitude','Latitude'])).head()
```

```python
saw.change_parameters_multiple_element_df('substation', sub_coords)
```

```python
# New sub locations centered around WA
(new_subs := saw.GetParametersMultipleElement(ObjectType='substation', ParamList=['SubNum','Longitude','Latitude'])).head()
```

```python
saw.SaveCase(pwb_path)
```
